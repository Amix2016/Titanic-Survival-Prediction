#### 1.2 Recognising Variables in Dataset <a id=3></a>

**Variable definitions in the Dataset**
* survival: Survival (0 = No, 1 = Yes)
* pclass: Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd)
* sex: Sex
* Age: Age in years
* sibsp: # of siblings / spouses aboard the Titanic
* parch: # of parents / children aboard the Titanic	
* ticket: Ticket number
* fare: Passenger fare
* cabin: Cabin number
* embarked: Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

**Additional variable descriptions to help us**

pclass: A proxy for socio-economic status (SES)
   * 1st = Upper
   * 2nd = Middle
   * 3rd = Lower
   
age: Age is fractional if less than 1. If the age is estimated, it is in the form of xx.5

sibsp: The dataset defines family relations in this way...

* Sibling = brother, sister, stepbrother, stepsister

* Spouse = husband, wife (mistresses and fiancs were ignored)

* parch: The dataset defines family relations in this way...
    * Parent = mother, father
    * Child = daughter, son, stepdaughter, stepson
    Some children travelled only with a nanny, therefore parch=0 for them.

Note: You can find these varaiables definitions in the data section of the "Titanic - Machine Learning from Disaster" dataset.

##### 3.1.1 Imputing Missing Values and Pre Processing <a id=48></a>

```python
train_df.corr()['Age']
```

```python
# Fill missing values in the 'Age' column using the median age of each passenger class

train_df['Age'] = train_df['Age'].fillna(train_df.groupby('Pclass')['Age'].transform('median'))
test_df['Age'] = test_df['Age'].fillna(train_df.groupby('Pclass')['Age'].transform('median'))
```

```python
# filling missing with mode
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0],inplace=True)
```

```python
train_df.isnull().sum()
```

```python
# replacing T viz Boat Deck to missing category
train_df['Cabin'].replace('T','U0',inplace=True)
```

##### Decks in Titanic Ships
* Boat Deck (54471257)
* A-Deck (59091145)
* B-Deck (93121176)
* C Deck (93261182)
* D Deck (91951224)
* E Deck (91741210)
* F Deck (88601205)
* G Deck (88481163)

```python
# Replace missing values in the 'Cabin' column with 'U0', extract the first character, and map the values to numeric categories
train_df['Cabin'] = train_df['Cabin'].fillna('U0').str[0].map({"A": 1, "B": 2, "C": 3,
                                                               "D": 4, "E": 5, "F": 6, "G": 7, "U": 8})
```

```python
# Extract titles from the 'Name' column and assign them to the 'Title' column
train_df['Title'] = train_df['Name'].str.split(',').str[1].str.split('.').str[0].str.strip()
```

```python
train_df['Title'].value_counts()
```

```python
# Replace 'Mlle' and "Ms" with Miss and other titles as 'Others' in the 'Title' column with the desired replacements
train_df['Title'].replace({'Mlle': 'Miss', 'Ms': 'Miss', 'Dr': 'Others', 'Rev': 'Others', 'Major': 'Others',
                           'Col': 'Others', 'the Countess': 'Others', 'Capt': 'Others', 'Sir': 'Others',
                           'Lady': 'Others', 'Mme': 'Others', 'Don': 'Others', 'Jonkheer': 'Others'}, inplace=True)
```

```python
train_df['Title'].value_counts()
```

```python
# Map the values in the 'Title' column to numerical values
train_df['Title'] = train_df['Title'].map({'Mr': 1, 'Miss': 2, 'Mrs': 3, 'Master': 4, 'Others': 5})
```

```python
# We find that multiple passengers had same ticket number
train_df.groupby('Ticket')['PassengerId'].count()
```

```python
# finding count of passengers on ticket
train_df['PassengersOnTicket'] = train_df.groupby('Ticket')['Ticket'].transform('count')
```

```python
# find per passenger ticket fare 
train_df['FarePerPassenger'] = train_df['Fare'] / train_df['PassengersOnTicket']
```

```python
# Calculate the size of the family for each passenger
train_df['Family'] = train_df['SibSp'] + train_df['Parch'] + 1
```

```python
# creating a category out of family size
def family_size(x):
    
    if x == 1:
        return 'Alone'
    elif x < 5:
        return 'Small'
    else:
        return 'Large'
```

```python
# apply above category to family column and create new feature Familytype
train_df['Familytype'] = train_df['Family'].apply(family_size)
```

```python
# to extract alphabets letters for the ticket 
def extract_string(i):
    if len(i) < 2:
        return '0'
    else:
        return i[0].strip()
```

```python
# Split the 'Ticket' column by whitespace and apply a function to extract the string part
train_df['Ticket'] = train_df['Ticket'].str.split().apply(extract_string)

# Remove the period and slash characters from the 'Ticket' column
train_df['Ticket'] = train_df['Ticket'].str.replace('.', '').str.replace('/', '')

# Convert the 'Ticket' column to uppercase
train_df['Ticket'] = train_df['Ticket'].str.upper()
```

```python
train_df['Ticket'].value_counts()
```

```python
# Calculate the count of each unique ticket value
ticket_counts = train_df['Ticket'].value_counts()

# Select the tickets that have a count less than 10 and assign them the label 'Others'
other_tickets = ticket_counts[ticket_counts < 10].index.to_list()
train_df.loc[train_df['Ticket'].isin(other_tickets), 'Ticket'] = 'Others'

# Display the count of each value in the 'Ticket' column
train_df['Ticket'].value_counts()
```

```python
train_df.head()
```

```python
# removing variables which have been transformed or new feature for the same created
train_df.drop(columns=['Fare','SibSp','Parch','Name','Family'],inplace=True)
```

##### A small p-value (typically less than 0.05) indicates that the data significantly deviate from a normal distribution, suggesting non-normality. 

#### Using t-test for 2 independent samples

The t-test is a parametric statistical test used to compare the means of two groups and make inferences about the population from which the samples were drawn. The t-test relies on several assumptions for its validity.
1. Independent observations: The observations within each group are assumed to be independent of each other. This means that the values in one group do not depend on or influence the values in the other group.
2. Normality: The distribution of the data within each group should approximately follow a normal distribution. This assumption is more critical when the sample sizes are small. However, the t-test can still be robust to violations of normality when the sample sizes are large. (Using CLT theorem when sample size is large this assumption can be ignored.)
3. Homogeneity of variances (equal variances): The variances of the data within each group should be approximately equal. This assumption is important when performing an independent samples t-test, where the variances of the two groups are assumed to be equal. Violations of this assumption can affect the accuracy of the t-test results.


##### To assess whether the variances of two groups are not equal, we will use Levene's Test:
The test's null hypothesis is that the variances are equal between the two groups.If the p-value is below a chosen significance level (e.g., 0.05), you can reject the null hypothesis and conclude that the variances are significantly different.

```python
# Code for conducting Levene's test for each numeric variable
for i in numeric_var:
    num_survived = train_df[train_df['Survived'] == 1][i]
    num_not_survived = train_df[train_df['Survived'] == 0][i]

    # Perform Levene's test
    statistic, p_value = stats.levene(num_survived, num_not_survived)

    # Print the results
    print("Levene's test statistic for {}:".format(i), statistic)
    print("P-value for {}:".format(i), p_value)
    print('-----------------')
```

##### Since P-value < 0.05, we can conclude that variances are significantly different.

##### Hypothesis :
**Null Hypothesis (H0)**: There is no significant difference in the average numeric variables between the survivors and non-survivors.

**Alternative Hypothesis (HA)**: There is a significant difference in the average numeric variables between the survivors and non-survivors.


```python
# Code for conducting independent t-tests for each numeric variable
for i in numeric_var:
    num_survived = train_df[train_df['Survived'] == 1][i]
    num_not_survived = train_df[train_df['Survived'] == 0][i]

    # Perform a t-test
    t_statistic, p_value = stats.ttest_ind(num_survived, num_not_survived, equal_var=False)

    # Print the results
    print("T-Statistic for {}:".format(i), t_statistic)
    print("P-Value for {}:".format(i), p_value)
    print('--------------------')
```

##### equal_var:
If True (default), perform a standard independent 2 sample test that assumes equal population variances. If False, perform Welchs t-test, which does not assume equal population variance.

#### Using Mann-Whitney U test
It is a non-parametric test which does not assume data distribution.

```python
# Code for conducting Mann-Whitney U tests for each numeric variable
for i in numeric_var:
    num_survived = train_df[train_df['Survived'] == 1][i]
    num_not_survived = train_df[train_df['Survived'] == 0][i]

    # Perform Mann-Whitney U test
    statistic, p_value = stats.mannwhitneyu(num_survived, num_not_survived)

    print("Mann-Whitney U test for {}:".format(i))
    print("Statistic for {}:".format(i), statistic)
    print("p-value for {}:".format(i), p_value)
    print('----------------')
```
